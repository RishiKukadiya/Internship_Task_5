```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic dataset.csv")
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

```python

df['Embarked'].value_counts()
```

```python

df['Sex'].value_counts()
```

```python
sns.pairplot(df, hue='Survived', diag_kind='kde', palette='Set1')
plt.show()
```

```python
df['Age'].hist(bins=30)
plt.title('Distribution of Age')
plt.show()
```

**Observation:**

**Age is right-skewed. Most passengers are young (20–40 years).**

```python
df['Embarked'].hist(bins=30)
plt.title('Distribution of Embarked')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.show()
```

**Observations:-**

**1st class passengers paid much higher fares compared to 3rd class**

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.show()
```

**Observation:**

**1. Higher fare-paying passengers have a higher survival rate.**

**2. Young and old passengers both paid lower fares on average.**

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.show()
```

**The median age of survivors is lower compared to non-survivors.**

**Young passengers (children and teenagers) had higher survival rates.**

**Very old passengers (>60 years) had lower survival rates**

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

**Observation**

**1st class passengers had the highest survival rate.**

**3rd class passengers had the lowest survival rate.**

**Being in lower class (3rd class) significantly reduced survival chances.**

**Ticket class (Pclass) had a strong impact on survival**

#🚢 Titanic Survival Analysis Report


**📋 Introduction**

This project analyzes the Titanic dataset to explore the factors that influenced passenger survival.
We focus on the relationships between Age, Fare, Pclass, and Sex with Survived status using visualization and statistical observations.


**📊 Exploratory Data Analysis (EDA)**

**🔹 Relationship between Age and Survived**

Younger passengers, especially children, had a higher chance of survival.

Older passengers (above 50 years) had lower survival rates.

Observation: Age played a significant role; survival favored the young.

**🔹 Relationship between Fare and Survived**

1. Passengers who paid higher fares were more likely to survive.

2. Low-fare paying passengers had lower survival rates.

3. Observation: Fare was positively correlated with survival chances.

**🔹 Impact of Pclass on Survived**

1. 1st class passengers had the highest survival rate.

2. 3rd class passengers faced the highest mortality.

3. Observation: Higher ticket class (Pclass 1) passengers were prioritized during rescue.

**🔹 Relationship between Gender (Sex) and Survived**

1. Females had a much higher survival rate than males.

2. Majority of males did not survive.

3. Observation: Sex was a strong predictor of survival due to rescue priority for women.

**✨ Summary of Key Findings**

1. Females, 1st class passengers, higher fare payers, and younger individuals had better survival chances.

2. Pclass, Sex, Fare, and Age are the most important features affecting survival.

3. Clear evidence of socio-economic factors (class and fare) influencing survival during the Titanic disaster.
